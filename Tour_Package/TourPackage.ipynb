# Converting Catogorical col to Numeric Col

```python
df['Occupation'].replace(to_replace={'Fe Male':'Female'},inplace=True)
df['MaritalStatus'].replace(to_replace={'Unmarried':'Single'},inplace=True)
df['TypeofContact'].replace(to_replace={np.nan:'Self Enquiry'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.dtypes
```

# Spliting data into X and Y

```python
X=df.drop('ProdTaken',axis=1)
Y=df['ProdTaken']
```

```python
X.head()
```

```python
Y.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.3, random_state = 101)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logi_model=LogisticRegression()
logi_model.fit(X_train,Y_train)
```

```python
print('Logistic Regression Model Score',logi_model.score(X_train,Y_train)*100)
```

## Confusion Matrix of LR

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
Y_pred=logi_model.predict(X_test)
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
print(classification_report(Y_test, Y_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
```

```python
accuracy
```

## Confusion Matrix of DTC

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
Y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
```
